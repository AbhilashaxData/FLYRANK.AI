## 2. The question: decision, action, cost of a wrong call

### Research Question

Which content pages should be reviewed first for refresh based on observable search and engagement signals?

### Decision

Determine which pages deserve priority for review.

### Action

A content reviewer uses the ranked recommendations to decide whether a page should be refreshed, expanded, monitored or left unchanged.

### Unit of analysis

One content page.

### Output

A ranked list of content pages with an opportunity score and supporting reason codes.

### Cost of a wrong recommendation

A false positive means reviewers spend time on pages that were not actually high priority.

A false negative means an important declining page is overlooked, potentially allowing further loss of traffic.

### Why data and machine learning help

Many search and engagement signals interact in ways that are difficult to capture using fixed rules. Machine learning can combine these observable signals to prioritise review candidates while supporting, rather than replacing, human decision making.

```python
research_question = "Which content pages should be reviewed first for refresh?"
unit_of_analysis="One content page"
output="Ranked review queue"
print("Research Question:", research_question)
print("Unit of Analysis:", unit_of_analysis)
print("Output:", output)
```
